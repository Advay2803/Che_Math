# Problem 1 Flow and Bifurcation

1. You are studying energy potential diagrams and come across the following equation:
$\frac{dx}{dt} = rx+x^3-x^5$. Let **r = -2** and use the code cell below to plot the flow figure. Be sure to label axes. Some quick hints:
- You will likely have to zoom into the figure to see the important points. Consider using ylim and xlim.
- It may help to draw an abline at y = 0. Look up the command "axhline".

```python
def dxdt(x,r):
    dxdt = r*x+x**3-x**5
    return dxdt

y_values = np.linspace(-3,3,100)

dydt_values = []
for m in y_values:
    dydt_values.append(dxdt(m,-2))
    
# Plot derivative vs y
plt.xlabel("x")
plt.ylabel("dx/dt")
plt.title("Flow on a line")
plt.plot(y_values,dydt_values);
plt.axhline(y=0, color="black", linestyle="--");
plt.plot([0],[0], color='green', marker = 'o', linestyle = 'none',markersize=12);
# plt.xlim(-0.5,0.5)
# plt.ylim(-3,3);
```

**It is observed that for the given polynomial ODE system only one real root (x=0) exists, rest all the four roots are imaginary. Thus the system has only one stable steady state value.**

**Only one steady state is available at x=0.The stability of this point is very fragile and a small deviation can lead to overshoot of the system.**

# Problem 2 Linearization

You're given the nonlinear, first-order, time-invariant ODE $\frac{dy}{dt} = sin(y)$.

1. In the space below, plot the flow diagram for the ODE. 

```python
def dydt(y):
    dydt = math.sin(y)
    return dydt

y = np.linspace(0,12.6,100)

dydt_values = []
for m in y:
    dydt_values.append(dydt(m))
    
plt.plot(y,dydt_values);
plt.plot([math.pi,3*math.pi],[0,0], color='green', marker = 'o',markersize=12);
plt.plot([0,2*math.pi,4*math.pi],[0,0,0], color='red', marker = 'o',markersize=12);
plt.xlabel("y")
plt.ylabel("$\\frac{dy}{dt}$");
```

2. In the space below, type out the equation for the linearized form of our ODE when it is linearized about the lowest, positive-valued, stable steadystate.

```python
def dydt(yd):
    dydt = math.cos(math.pi)*(yd-math.pi)
    return dydt

y = np.linspace(1.5,5,100)

dydt_values = []
for m in y:
    dydt_values.append(dydt(m))
    
plt.axhline(y=0, color="black", linestyle="-")
plt.plot([math.pi],[0], color='green', marker = 'o',markersize=10);
plt.plot(y,np.sin(y),'r-.',label = "Original model",markersize=8)
plt.plot(y,dydt_values,label = "Linearized model",markersize=8);
plt.xlabel("y")
plt.ylabel("$\\frac{dy}{dt}$");
plt.legend();
```

3. We want to compare the responses of our linearized model to our original nonlinear model. On a single figure, plot the time response of both models when starting at an initial conditions of $y_0 = 3$. Clearly indicate which lines represent which ODEs.

```python
from scipy.integrate import odeint

def dydt1(yd,t):
    dydt1 = math.cos(math.pi)*(yd-math.pi)
    return dydt1

timesteps = np.linspace(0,10,100)

y0 = 3.0
yd = odeint(dydt1, y0, timesteps)

def dydt(y,t):
    dydt = math.sin(y)
    return dydt

y = odeint(dydt, y0, timesteps)

plt.rcParams.update({'font.size': 14})  
plt.xlabel("time")
plt.ylabel("y")
plt.plot(timesteps, yd,'r-',label="linear");
plt.plot(timesteps, y,'b--',label="nonlinear");
plt.legend();
```
